```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

import pandas_market_calendars as mcal

data_folder = './data/'
```

```python
df = pd.read_csv(data_folder + 'NASDAQ100_m.csv')
df
```

```python
df.replace({-9999: np.nan}, inplace=True)
df['Time'] = df['Time']+'+00:00'
df.interpolate(method='linear',limit_direction='both', inplace=True)
df.set_index('Time', inplace=True)
df.head()
```

```python
date = mcal.get_calendar('XKRX')
```

```python
early = date.schedule(start_date='2019-07-02', end_date='2020-01-14')
df.index = pd.to_datetime(df.index)

df = df.resample('1min').first().loc[mcal.date_range(early, frequency='1min')].interpolate(method='linear', limit_direction='both')
df
```

```python
df.to_csv(data_folder + 'KOSPI200_m.csv')
```
